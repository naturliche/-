######  分析独立IP数

```python
f = open(r"C:\Users\natur\Desktop\work\log_anay\last-1000-log", "r") 
arr = {}      #用字典来存储IP跟访问次数
ip = []
v1_ip = []     #非独立ip
for line in f:
    lines = line.split()
    v_ip = lines[0]
    v1_ip.append(v_ip)
    if not v_ip in ip:
        ip.append(v_ip)
        
for item in ip:
    key = item
    value = v1_ip.count(item)
    arr.setdefault(key,value)

b_sort = sorted(arr.items(),key = lambda x:x[1],reverse = True)
print(b_sort)
f.close()
```

###### url

```python
f = open(r"C:\Users\natur\Desktop\work\log_anay\last-1000-log", "r")
arr = {}
url = []
v1_url = []
for line in f:
    lines = line.split('/',3)
    v_url = lines[3]
    a = v_url.split('?')
    b = a[0]
    v1_url.append(b)
    if not b in url:
        url.append(b)
        
for item in url:
    key = item
    value = v1_url.count(item)
    arr.setdefault(key,value)

b_sort = sorted(arr.items(),key = lambda x:x[1],reverse = True)
print(b_sort)
f.close()
```
